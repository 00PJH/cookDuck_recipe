### 데이터 합치기


```python
import pandas as pd
folder_path = '../data/'
file_name_first = 'recipe_'
file_name_second = 'data'
file_name_third = 'process'
file_name_last = '_cleaned.csv'
output_file_name = 'data_process_combined.csv'
```

```python
df_prc = pd.read_csv(folder_path + file_name_first + file_name_third + file_name_last ,
                    encoding='cp949',
                    engine='python',
                    on_bad_lines='warn')
df_data = pd.read_csv(folder_path + file_name_first + file_name_second + file_name_last,
                    encoding='cp949',
                    engine='python',
                    on_bad_lines='warn')

df_prc.head()
```

```python
df_data.head()  
```

```python
df_merged = pd.merge(
        df_data,
        df_prc[['레시피 코드', '요리설명']],  # process 파일에선 필요한 열만 선택
        on='레시피 코드',
        how='left'  # data 파일 기준으로 합치기
    )
```

```python
df_merged.head()
```

```python
# 4-1. 맨 뒤에 붙은 '요리설명' 열을 뽑아냅니다. (pop)
    #      (이때 df_merged에서는 '요리설명' 열이 삭제됩니다)
col_to_move = df_merged.pop('요리설명')

# 4-2. '요리설명'을 삽입할 위치, 즉 '요약' 열의 인덱스 번호를 찾습니다.
insert_location = df_merged.columns.get_loc('요약')

# 4-3. 찾은 위치(insert_location)에 '요리설명' 열을 삽입(insert)합니다.
df_merged.insert(insert_location, '요리설명', col_to_move)

# ----------------------------------------------------
# 5. '레시피 코드'로 오름차순 정렬
# ------------------------------------------------
df_sorted = df_merged.sort_values(by='레시피 코드')
```

```python
df_sorted.head()
```

```python
df_sorted.to_csv(folder_path + output_file_name, index=False, encoding='utf-8-sig') 
```
