## 1. 요리 과정 csv파일 전처리

```python
import pandas as pd

file_path = '../data/'
file_name = 'recipe_process.csv'

df_prc = pd.read_csv(
    file_path + file_name,
    encoding='cp949',
    engine='python',
    on_bad_lines='warn' 
)

df_prc.head(10)

df_prc = df_prc.drop(columns=['과정 이미지 URL'])
df_prc.head(10)

df_prc_sorted = df_prc.sort_values(by=['레시피 코드', '요리설명순서']).reset_index(drop=True)
df_prc_sorted.head(10)
```

```python
# 1. 각 요리설명 앞에 '순서번호. '를 붙여주는 함수를 정의합니다.
def format_steps(group):
    # 예: "1. 재료를 손질한다.\n2. 냄비에 넣고 끓인다." 와 같은 형식으로 만듭니다.
    # str(x['요리설명순서'])를 사용해 숫자형 순서를 문자열로 변환합니다.
    formatted_list = group.apply(
        lambda x: str(x['요리설명순서']) + '. ' + x['요리설명'],
        axis=1
    )
    return '\n'.join(formatted_list)

# 2. '레시피 코드'로 그룹화한 뒤, 위에서 만든 함수를 각 그룹에 적용합니다.
combined_steps = df_prc_sorted.groupby('레시피 코드').apply(format_steps)

# 3. 결과를 새로운 데이터프레임으로 만듭니다.
# '요리설명'이라는 열 이름을 지정해줍니다.
df_combined = pd.DataFrame(combined_steps, columns=['요리설명']).reset_index()

# 최종 결과 확인
df_combined.head()
```

```python
# df_combined 데이터프레임에서 '레시피 코드'가 1인 행을 찾습니다.
# .loc[행 조건, 열 이름]
recipe_description = df_combined.loc[df_combined['레시피 코드'] == 1, '요리설명'].values[0]

# print() 함수를 사용하면 줄바꿈(\n)이 적용되어 보기 좋게 출력됩니다.
print(recipe_description)
```

```python
# 저장
output_file_name = 'recipe_process_cleaned.csv'
df_combined.to_csv(file_path + output_file_name, index=False, encoding='cp949')
```
